LOAD

```python
import pandas as pd
import os

# Define the raw data path
raw_path = "../data/raw/"

# Load your CSVs
file1 = pd.read_csv(os.path.join(raw_path, "index.csv"))
file2 = pd.read_csv(os.path.join(raw_path, "stocks.csv"))

print("Index file shape:", file1.shape)
print("Stocks file shape:", file2.shape)
```

Clean column names + convert date columns

```python
# Clean column names
file1.columns = file1.columns.str.strip().str.lower()
file2.columns = file2.columns.str.strip().str.lower()

# Convert date columns to datetime
file1['date'] = pd.to_datetime(file1['date'])
file2['date'] = pd.to_datetime(file2['date'])

print(file1.head())
print(file2.head())
```

Merge with NIFTY (index) data

```python
# For now, skip merge and just use stock data
merged = file2.copy()

print("Merged file shape:", merged.shape)
merged.head()
```

```python
merged = merged.sort_values(by=['date'])
merged = merged.fillna(method='ffill').fillna(method='bfill')
merged.info()
```

```python
processed_path = "../data/processed/"
os.makedirs(processed_path, exist_ok=True)

merged.to_csv(os.path.join(processed_path, "merged_clean.csv"), index=False)
print("✅ Saved cleaned dataset at:", os.path.join(processed_path, "merged_clean.csv"))
```
